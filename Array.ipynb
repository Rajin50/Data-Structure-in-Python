```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence,beats):
  for i in beats:
    if i == 1:
      temp= sequence[len(sequence) - 1]
      for j in range(len(sequence) - 1, 0, -1):
          sequence[j] = sequence[j-1] 
    else:
      pass
    sequence[0] = temp
  return sequence 


# Test 02: Discard Cards
def discardCards(cards,number):
  j=0

  while j<len(cards):
    if cards[j] == number:
      for i in range(j, len(cards)-1):
        cards[i] = cards[i+1]
      cards[len(cards)-1] = 0
      j -= 1

    j += 1
  return cards


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  for i in range(0, len(pokemon_1)):
    if pokemon_1[i] == None:
      pokemon_1[i] = 0
  for j in range(0, len(pokemon_2)):
    if pokemon_2[j] == None:
      pokemon_2[j] = 0

  new_array = [None] * len(pokemon_1)

  for x in range(0, len(pokemon_1)):
    new_array[x] = pokemon_1[x] + pokemon_2[len(pokemon_2) - 1 - x]

  return np.array(new_array) 


# Test 04: Balance your Salami
def balanceSalami(salami):
  total = 0
  for i in range(0, len(salami)):
     total += salami[i]

  s = 0
  for j in range(0, len(salami)):
    if total != s:
      total -= salami[j]   
      s += salami[j]
      
  if s == total:
    return True 
  else:
    return False


# Test 05: Protect Salami
def protectSalami(salami):
  new1 = np.array([0] * len(salami))
  for i in range(0, len(salami)):
    new1[salami[i]] += 1

  large = 0
  for num in new1:
    if large < num:
      large = num

  new2 = np.array([0] * (large+1))
  for j in range(0, len(new1)):
    new2[new1[j]] += 1

  v = 0
  for value in range(2, len(new2)):
    if new2[value] > 1:
      v += 1
  
  if v == 0:
    return False
  else:
    return True 


# Test 06: Odd Even Wave
def waveYourFlag(arr):
  new = [0]*len(arr)
  even = 2
  odd = 1

  if arr[0] % 2 == 0:
    new[0] = arr[0]
    for i in range(1, len(arr)):
      if arr[i] % 2 == 0:
        new[even] = arr[i]
        even += 2

      else:
        new[odd] = arr[i]
        odd += 2

  else:
    new[0] = arr[0]
    for j in range(1, len(arr)):
      if arr[j] % 2 != 0:
        new[even] = arr[j]
        even += 2

      else:
        new[odd] = arr[j]
        odd += 2

  return np.array(new)

```
