```python

class Patient:
  def __init__(self, id, name, age, bg, next = None, prev = None):
    self.id = id
    self.name = name
    self.age = age
    self.bg = bg
    self.next = next
    self.prev = prev

class WRM:
  def __init__(self):
    self.dummyh = Patient(None, None, None, None)
    self.dummyh.next = self.dummyh
    self.dummyh.prev = self.dummyh
    self.tail = self.dummyh

  def RegisterPatient(self, id, name, age, bg):
    new_patient = Patient(id, name, age, bg)
    self.tail.next = new_patient
    new_patient.prev = self.tail
    new_patient.next = self.dummyh
    self.dummyh.prev = new_patient
    self.tail = self.tail.next

    return "Patient Successfully Registered"

  def ServePatient(self):
    tail = self.dummyh.next
    if tail == self.dummyh:
      print("No Patient to be served")
    else:
      print(tail.name, "is served")
      temp = self.dummyh.next.next
      self.dummyh.next = temp
      temp.prev = self.dummyh
      tail.next = None
      tail.prev = None

  def CancelAll(self):
    self.dummyh.next = None
    self.dummyh.prev = None
    self.dummyh.next = self.dummyh
    self.dummyh.prev = self.dummyh
    self.tail = self.dummyh
    print("All appointments cancelled")

  def ShowAllPatient(self):
    tail = self.dummyh.next

    if tail.id == None:
      print("No Patient ID")

    while True:
      if tail.id == None:
        break

      print(tail.id)
      tail = tail.next

  def CanDoctorGoHome(self):
    tail = self.dummyh.next

    while True:
      if tail.id == None:
        return True
      return False
      tail = tail.next

w1 = WRM()
while True:
  print("1. RegisterPatient()")
  print("2. ServePatient()")
  print("3. CancelAll()")
  print("4. ShowAllPatient()")
  print("5. CanDoctorGoHome()")
  print("6. Exit")
  print("===============================================================")

  x = int(input("Enter Your Choice: "))
  if x == 6:
    print("Thank you for using waiting room management system")
    break

  elif x > 6 or x < 0:
    print("No such option")

  elif x == 1:
    id = input("Enter Your ID: ")
    name = input("Enter Your Name: ")
    age = input("Enter Your Age: ")
    bg = input("Enter Your Blood Group: ")

    print(w1.RegisterPatient(id, name, age, bg))

  elif x == 2:
    w1.ServePatient()

  elif x == 3:
    w1.CancelAll()

  elif x == 4:
    w1.ShowAllPatient()

  elif x == 5:
    c = w1.CanDoctorGoHome()
    if c == True:
      print("Yes")
    else:
      print("No")

```
